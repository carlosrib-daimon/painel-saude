#### Novo db com leitos

```python
df = pd.read_csv("rlEstabComplementar202502.csv", encoding='latin1', sep=';')
df.head()
```

```python
result = df.groupby('CO_UNIDADE').agg(
    QT_EXIST=('QT_EXIST', 'sum'),
    QT_SUS=('QT_SUS', 'sum')
).reset_index()

result.head()
```

### Juntando Informações

```python
df_temp = result[['CO_UNIDADE', 'QT_EXIST', 'QT_SUS']].merge(
    df_unid[['CO_UNIDADE', 'CO_MUNICIPIO_GESTOR', 'CO_NATUREZA_JUR', 'CO_MOTIVO_DESAB']],
    on='CO_UNIDADE',
    how='left'
)


df_new = df_temp.merge(
    df_regioes,
    left_on='CO_MUNICIPIO_GESTOR',
    right_on='Codigo Municipio',
    how='left'
)

df_new.head()
```
